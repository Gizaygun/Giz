```python
import numpy as np
import pandas as pd
from datasets import Dataset
from transformers import pipeline

# --- 1. Hugging Face Model Yükleme ---
# GPT-2 modeli hızlı ve etkili bir seçimdir; daha büyük modeller gerekirse değiştirilebilir.
text_generator = pipeline("text-generation", model="gpt2")

# --- 2. Veri Simülasyonu ---
num_students = 2000
data = {
    "Student_ID": np.arange(1, num_students + 1),
    "Age": np.random.randint(18, 30, num_students),
    "Class_Level": np.random.choice(["Freshman", "Sophomore", "Junior", "Senior"], num_students),
    "Interest": np.random.choice(["Math", "Science", "Art", "History"], num_students),
    "Exam_Score": np.random.randint(50, 100, num_students),
    "Study_Hours": np.random.uniform(1, 10, num_students).round(2),
    "Engagement_Score": np.random.randint(1, 101, num_students),
    "Content_Viewed": np.random.randint(1, 50, num_students),
}

# DataFrame'den Hugging Face Dataset'e dönüştürme
df = pd.DataFrame(data)
dataset = Dataset.from_pandas(df)

# --- 3. Doğal Dil Cümlesi Üretimi ---
def generate_sentence(row):
    prompt = (f"A student interested in {row['Interest']} spends about {row['Study_Hours']} hours studying, "
              f"has an engagement score of {row['Engagement_Score']}, and viewed {row['Content_Viewed']} materials. "
              "Describe their learning behavior on the platform.")
    result = text_generator(prompt, max_length=50, num_return_sequences=1)
    return result[0]["generated_text"]

# Yeni sütun için cümleler oluşturma
dataset = dataset.map(lambda row: {"Behavior_Description": generate_sentence(row)})

# --- 4. Veri Setini Kaydetme ---
dataset.save_to_disk("enhanced_dataset")
print("Zenginleştirilmiş veri seti kaydedildi: enhanced_dataset")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity

# --- 1. Veri Setini Yükleme ---
dataset_path = "/content/enhanced_dataset (1).csv"  # Dosya yolunu belirtin
dataset = pd.read_csv(dataset_path)

# --- 2. Eksik Verileri Kontrol Etme ve Doldurma ---
# Sayısal veriler için ortalama ile doldurma
numerical_cols = dataset.select_dtypes(include=['float64', 'int64']).columns
dataset[numerical_cols] = dataset[numerical_cols].fillna(dataset[numerical_cols].mean())

# Kategorik veriler için mod (en yaygın değer) ile doldurma
categorical_cols = dataset.select_dtypes(include=['object']).columns
for col in categorical_cols:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])

# --- 3. One-Hot Encoding Kategorik Veriler ---
dataset = pd.get_dummies(dataset, columns=['Class_Level', 'Interest'], drop_first=True)

# --- 4. Sayısal Verileri Normalleştirme ---
scaler = MinMaxScaler()
if 'Study_Hours' in dataset.columns:
    dataset['Study_Hours'] = scaler.fit_transform(dataset[['Study_Hours']])

# --- 5. Eğitim ve Test Verisine Ayırma ---
train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)

# --- 6. Özellikler ve Hedef Değişkenlerin Seçimi ---
# Yalnızca sayısal sütunları seçiyoruz
numerical_features = train_data.select_dtypes(include=['float64', 'int64', 'uint8']).columns

X_train = train_data[numerical_features].drop(columns=['Exam_Score', 'Student_ID'], errors='ignore')
y_train = train_data['Exam_Score']  # Hedef değişken

X_test = test_data[numerical_features].drop(columns=['Exam_Score', 'Student_ID'], errors='ignore')
y_test = test_data['Exam_Score']  # Test hedef değişkeni

# --- 7. Öğrenciler Arasındaki Benzerliği Hesaplama (Cosine Similarity) ---
similarity_matrix = cosine_similarity(X_train)

# --- 8. İçerik Tabanlı Öneri Fonksiyonu ---
def recommend_for_student(student_id, similarity_matrix, X_train, top_n=5):
    student_index = student_id - 1  # ID'yi sıfır tabanlı yapıyoruz
    if student_index >= len(similarity_matrix):
        raise ValueError(f"Geçersiz öğrenci ID: {student_id}. Veri setinde {len(similarity_matrix)} öğrenci var.")

    similar_students = similarity_matrix[student_index]

    # Benzer öğrenciler sıralanır
    similar_students_idx = similar_students.argsort()[-top_n:][::-1]

    recommended_courses = X_train.iloc[similar_students_idx]
    return recommended_courses

# --- 9. Öneri Yapma (Örnek: Öğrenci 10 için) ---
student_id = 10  # Örnek öğrenci ID
try:
    recommended_courses = recommend_for_student(student_id, similarity_matrix, X_train)
    print(f"Öğrenci {student_id} için önerilen dersler:\n{recommended_courses}")
except ValueError as e:
    print(e)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

# --- 1. Veri Setini Yükleme ---
dataset_path = "/content/enhanced_dataset (1).csv"  # Dosya yolunu belirtin
dataset = pd.read_csv(dataset_path)

# --- 2. Eksik Verileri Kontrol Etme ve Doldurma ---
# Sayısal veriler için ortalama ile doldurma
numerical_cols = dataset.select_dtypes(include=['float64', 'int64']).columns
dataset[numerical_cols] = dataset[numerical_cols].fillna(dataset[numerical_cols].mean())

# Kategorik veriler için mod ile doldurma
categorical_cols = dataset.select_dtypes(include=['object']).columns
for col in categorical_cols:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])

# --- 3. One-Hot Encoding Kategorik Veriler ---
dataset = pd.get_dummies(dataset, columns=['Class_Level', 'Interest'])

# --- 4. Sayısal Verileri Normalleştirme ---
scaler = MinMaxScaler()
dataset['Study_Hours'] = scaler.fit_transform(dataset[['Study_Hours']])

# --- 5. Özellik Mühendisliği: Yeni Özellikler Ekleyelim ---
dataset['Total_Engagement'] = dataset['Study_Hours'] * dataset['Engagement_Score']  # Etkileşimi başarıyla birleştirme

# --- 6. Eğitim ve Test Verisine Ayırma ---
train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)

# --- 7. Özellikler ve Hedef Değişkenlerin Seçimi ---
X_train = train_data[['Study_Hours', 'Engagement_Score', 'Total_Engagement']]  # Kullanılacak özellikler
y_train = train_data['Exam_Score']  # Hedef değişken

X_test = test_data[['Study_Hours', 'Engagement_Score', 'Total_Engagement']]  # Test verisi özellikleri
y_test = test_data['Exam_Score']  # Test verisi hedef değişkeni

# --- 8. Content-Based Filtering (İçerik Tabanlı Filtreleme) ---
similarity_matrix = cosine_similarity(X_train)

def recommend_for_student_content_based(student_id, similarity_matrix, X_train, top_n=5):
    student_index = student_id - 1  # ID'yi sıfır tabanlı yapıyoruz
    similar_students = similarity_matrix[student_index]
    similar_students_idx = similar_students.argsort()[-top_n:][::-1]
    recommended_courses = X_train.iloc[similar_students_idx]
    return recommended_courses

# --- 9. Collaborative Filtering (İşbirlikçi Filtreleme) ---
knn = KNeighborsRegressor(n_neighbors=6, metric='cosine')
knn.fit(X_train, y_train)
y_pred_collaborative = knn.predict(X_test)

def recommend_for_student_collaborative(student_id, knn_model, X_train, top_n=5):
    distances, indices = knn_model.kneighbors(X_train)
    similar_students_idx = indices[student_id - 1][1:]
    recommended_courses = X_train.iloc[similar_students_idx]
    return recommended_courses

# --- 10. Öneri Yapma (Örnek: Öğrenci 10 için) ---
student_id = 10
print(f"Öğrenci {student_id} için Content-Based öneriler:\n", recommend_for_student_content_based(student_id, similarity_matrix, X_train))
print(f"Öğrenci {student_id} için Collaborative öneriler:\n", recommend_for_student_collaborative(student_id, knn, X_train))

# --- 11. Model Değerlendirmesi (RMSE) ---
# Content-Based Model için regresyon
content_based_model = LinearRegression()
content_based_model.fit(X_train, y_train)
y_pred_content_based = content_based_model.predict(X_test)

# Collaborative Model için regresyon
collaborative_model = LinearRegression()
collaborative_model.fit(X_train, y_train)
y_pred_collaborative = collaborative_model.predict(X_test)

rmse_content_based = np.sqrt(mean_squared_error(y_test, y_pred_content_based))
rmse_collaborative = np.sqrt(mean_squared_error(y_test, y_pred_collaborative))

print("Content-Based RMSE:", rmse_content_based)
print("Collaborative RMSE:", rmse_collaborative)

# --- 12. Model İyileştirmeleri: Parametre Ayarlamaları ---

# KNN için GridSearchCV ile parametre tuning
param_grid = {'n_neighbors': [5, 10, 15], 'metric': ['cosine', 'euclidean']}
grid_search = GridSearchCV(KNeighborsRegressor(), param_grid, cv=3, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# En iyi parametreleri ve modeli elde etme
best_knn_model = grid_search.best_estimator_
y_pred_best = best_knn_model.predict(X_test)

# İyileştirilmiş modelin doğruluğunu değerlendirme
rmse_best = np.sqrt(mean_squared_error(y_test, y_pred_best))

print("Best KNN Model RMSE:", rmse_best)
print("Best KNN Model Parametreleri:", grid_search.best_params_)

```

```python
from IPython.core.display import display, HTML

# HTML ve CSS Kodlarını Tek Sayfa Olarak Hazırlıyoruz
html_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Student Recommendation System</title>
    <style>
        /* Genel stil */
        body {
            font-family: 'Arial', sans-serif;
            margin: 0;
            padding: 0;
            background: linear-gradient(to right, #00BFFF, #FF6347); /* Parlak mavi ve parlak turuncu geçiş */
            color: #fff;
            min-height: 100vh;
            display: flex;
            flex-direction: column;
            justify-content: center;
            align-items: center;
            text-align: center;
        }

        /* Başlık */
        h1 {
            font-size: 4.5em;
            font-weight: bold;
            margin: 0 0 20px 0;
            color: #fff;
            text-shadow: 3px 3px 6px rgba(0, 0, 0, 0.3);
        }

        /* Görsel stili */
        .hero-image {
            width: 100%;
            max-width: 700px;
            height: auto;
            border-radius: 15px;
            box-shadow: 0 0 15px rgba(0, 0, 0, 0.5);
            margin: 20px 0;
        }

        /* Form stili */
        form {
            background-color: rgba(0, 0, 0, 0.6);
            padding: 30px;
            border-radius: 10px;
            box-shadow: 0 0 20px rgba(0, 0, 0, 0.7);
            width: 100%;
            max-width: 500px;
            text-align: center;
        }

        /* Input ve buton */
        input[type="text"], button {
            padding: 15px;
            border-radius: 5px;
            border: none;
            width: 80%;
            margin: 10px 0;
            font-size: 1.3em;
        }

        /* Giriş kutusu */
        input[type="text"] {
            background-color: #fff;
            color: #333;
            text-align: center;
        }

        /* Buton */
        button {
            background-color: #FF6347;
            color: #fff;
            cursor: pointer;
            transition: background-color 0.3s ease;
        }

        button:hover {
            background-color: #FF4500;
        }

        /* İçerik liste stili */
        ul {
            list-style-type: none;
            padding: 0;
            margin: 0;
            font-size: 1.2em;
        }

        /* Liste öğeleri */
        li {
            background-color: #00BFFF;
            padding: 15px;
            margin: 10px 0;
            border-radius: 5px;
            text-align: center;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.2);
        }

        li:hover {
            background-color: #1E90FF;
        }

        footer {
            position: absolute;
            bottom: 20px;
            color: #fff;
            font-size: 1em;
            text-align: center;
        }
    </style>
</head>
<body>
    <h1>Student Recommendation System</h1>
    <img src="/mnt/data/A_visually_striking_and_professional_image_represe.png" alt="Recommendation Hero Image" class="hero-image">

    <form id="recommendation-form">
        <label for="student_id">Enter Your Student ID:</label>
        <input type="text" id="student_id" name="student_id" required>
        <button type="submit">Get Recommendation</button>
    </form>

    <h2>Recommended Courses</h2>
    <ul id="course-list">
        <!-- Dynamic recommendation will appear here -->
    </ul>

    <footer>
        <p>© 2024 Student Recommendation System | Designed for You</p>
    </footer>

    <script>
        // Recommendations based on student ID
        const courses = {
            1: ["Mathematics - 85", "Science - 92", "Social Studies - 78"],
            2: ["Chemistry - 88", "Biology - 90", "Physics - 80"],
            3: ["Art - 95", "History - 88", "Geography - 84"]
        };

        document.getElementById("recommendation-form").addEventListener("submit", function(event) {
            event.preventDefault();
            const studentId = document.getElementById("student_id").value;
            const courseList = document.getElementById("course-list");
            courseList.innerHTML = "";  // Clear previous recommendations

            if (courses[studentId]) {
                courses[studentId].forEach(function(course) {
                    const li = document.createElement("li");
                    li.textContent = course;
                    courseList.appendChild(li);
                });
            } else {
                const li = document.createElement("li");
                li.textContent = "No recommendations found for this student.";
                courseList.appendChild(li);
            }
        });
    </script>
</body>
</html>
"""

# Display HTML content in Google Colab
display(HTML(html_code))

# Save HTML content to a file
with open('/content/student_recommendation_system.html', 'w') as f:
    f.write(html_code)

# Download the HTML file
from google.colab import files
files.download('/content/student_recommendation_system.html')
```
